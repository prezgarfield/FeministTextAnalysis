## Tokenizing

Last week, what made things so hard is that we were learning about something called Regular Expressions. Regular expressions are a way of referring to text so that you can search for and manipulate strings of characters. What we did last week was use Regular Expressions to create a way of chopping one long string of text into smaller segments that the computer could use. 

That's a lot of work to do every time you want to do it! So, instead, we rely on the work that others have already done to make the task a little simpler. In the Natural Language Toolkit, there is a "module" called "tokenize." This means that we can use the work by linguists to chop up the text we want to work with in ways that are most frequently used. In this case, we'll tokenize _words_ and _sentences_. Pay attention as you go along to what is and what is not included in that process. 


### What's happening here? 

In the following example, we are taking the whole string of text in quotes and creating a placeholder, or _variable_ for it. In other words, this: "Hello Mr. Smith, how are you doing today? The weather is great, and city is awesome. The sky is pinkish-blue. You shouldn't eat cardboard." becomes the same thing as the variable "text." Two things happen in that first 2 lines. First, the whole string is not represented by the word "text." Second, the word "text" is now something that we can do something to because the computer knows that the word "text" now represents a specific _type_ of data: a string. 

Once the string of characters is represented by the word text, we can now do things to it. In this case, we're going to tell the computer to go find the short bit of repeatable code in the NLTK package (sent_tokenize) and do that to the variable _text_. The output of that activity will be stored with another variable: tokenized_sent.

When something works in Python, you may not always know, because Python "succeeds silently." That is to say, that unless you explicitly tell the computer to report its results, it won't show its work. For that reason, we need to tell the computer to show its results using the Python function: print(). Putting the variable inside the parentheses. 

```python
text="""Hello Mr. Smith, how are you doing today? The weather is great, and city is awesome.
The sky is pinkish-blue. You shouldn't eat cardboard."""
tokenized_sent=sent_tokenize(text)
print(tokenized_sent)
```

## Frequency Distribution
Once you have a text divided into smaller segments, we can count and manipulate those pieces by searching, counting, combining, measuring, clustering, and predicting. Obviously, the segment above is brief enough that you might be able to detect repetitions of words or their similiarity. 

A frequency distribution shows you the frequency with which a word is distributed across a sample. To do this we use the FreqDist module and we tell it to run those calculations using the data that we stored with the variable "tokenized_word." When we print the outcome of that process, which we represent as fdist, we then print it. But what we get a verbal answer. We can also print them out. We can also use this process to filter our results. For example, what are the 2 monst common terms? 

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokenized_word)
print(fdist)
```

# Stopwords
If you look closely at the distribution above, what you'll see is that the words that happen most frequently are words that are semantically necessary but they hold very little meaning comparatively. In other words, their frequency is so high that they can make it hard to interpret your results. There are a number of ways that we can deal with this. One is to have a list of those words that we determine happen so frequently in the language we are working with, and then we can simply strip those words out of the set so that we can see words that are more meaningful. 

In the code below, we're drawing a list of what we call "stopwords" that comes with NLTK, and we're filtering those out of our dataset. So first, we are going to pull in those English language stop words from the NLTK package and print them out so that you can see what those are. Next, we'll create a small process or loop to do a repetitive task across all the words in our sample.

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
print(stop_words)
```

In the next process, we're turning the text sample from the beginning of the notebook into two different sets of words. First, we go through each sentence and check each word against the stop word list. When the word matches one in the stopword list, we should remove it. When the word doesn't match one of the words in the stopword list, add it to the filtered_sent variable. 

In a markdown file below, try to explain what is printed out in the output cell. 

```python
filtered_sent=[]
for w in tokenized_word:
    if w not in stop_words:
        filtered_sent.append(w)
print("Tokenized Sentence:",tokenized_sent)
print("Filterd Sentence:",filtered_sent)
```

What do we see in "Tokenized Sentences" and "Filtered Sentence" in the printed output? 

**RG -- With stop words, we see the first significant "deformation" of the text.  We are saying that certain words are "noise" for our purposes and thus must be eliminated.

# Lexicon Normalization 

What do we do about words that mean very similar things and have similar root words, but they are different "strings" of characters in the text? Does it make sense to treat each instance individually, or should we combine them? There's a process that we use in preparing texts for computational analysis that combines words that have similar or the same meanings but that differ because of their tense, number, or possession, for example. 

```python
## Stemming
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps=PorterStemmer()

stemmed_words=[]
for w in filtered_sent:
    stemmed_words.append(ps.stem(w))

print("Filtered Sentence:",filtered_sent)
print("Stemmed Sentence:",stemmed_words)
```

# Lemmatization

When you lemmatize your corpus, you strip a word down to its "lemma" or stem. It transforms the word in order to combine similar "sense" or "meanings." Adverbs and adjectives, for example are words that are useful to lemmatize because they are similar and frequent enough to be considered "noise" in the text, but at the same time, they are not the same word root. The example from the lesson is "good" and "better." 

As the tutorial points out, stemming doesn't work because these words don't have the same root, and this process requires looking words up in a lexicon, or dictionary. The word "dictionary" will be used to mean several things as we proceed. It's good to check to make sure you're clear on which meaning of the word is intended. 

```python
# Lexicon Normalization
# performing stemming and Lemmatization

from nltk.stem.wordnet import WordNetLemmatizer
lem=WordNetLemmatizer()

from nltk.stem.porter import PorterStemmer
stem=PorterStemmer()

word="flying"
print("Lemmatized Word:",lem.lemmatize(word,"v"))
print("Stemmed Word:",stem.stem(word))
```

# Text Classification

The data in the following section can be found in the Week 5 repository on GitHub, but it was originally used in a Kaggle competition. It includes training and testing collections of words. 

```python
#import pandas
import pandas as pd

data=pd.read_csv('train.tsv', sep='\t')

data.head()
```

```python
data.info()
```

```python
data.Sentiment.value_counts()
```

```python
Sentiment_count=data.groupby('Sentiment').count()
plt.bar(Sentiment_count.index.values, Sentiment_count['Phrase'])
plt.xlabel('Review Sentiments')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer
#tokenizer to remove unwanted elements from out data like symbols and numbers
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(lowercase=True,stop_words='english',ngram_range = (1,1),tokenizer=token.tokenize)
text_counts = cv.fit_transform(data['Phrase'])
                        
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    text_counts, data['Sentiment'], test_size=0.3, random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
#Import scikit-learn metrics module for accuracy calcultion
from sklearn import metrics
#Model Generation Using Mutinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
predicted = clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
text_tf=tf.fit_transform(data['Phrase'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    text_tf, data['Sentiment'], test_size=0.3, random_state=123)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
#Model Generation Using Multinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
predicted = clf.predict(X_test)
print("MultinomialNB Accuracy:", metrics.accuracy_score(y_test, predicted))
```
